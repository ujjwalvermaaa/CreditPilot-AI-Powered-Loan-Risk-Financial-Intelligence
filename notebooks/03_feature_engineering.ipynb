## Next Step

 Model Training & Evaluation

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
df = pd.read_csv("../data/raw/creditpilot_dataset.csv")
df.head()
```

## Recalculate (ensuring consistency)

```python

df["emi_to_income_ratio"] = df["emi"] / (df["annual_income"] / 12)
df["loan_to_income_ratio"] = df["loan_amount"] / df["annual_income"]
df["debt_to_income_ratio"] = df["total_debt"] / df["annual_income"]

# Risk-related engineered feature
df["credit_risk_score"] = (
    (850 - df["credit_score"]) * 0.4 +
    df["debt_to_income_ratio"] * 100 * 0.3 +
    df["credit_utilization_ratio"] * 100 * 0.3
)
```

## Encoding Categorical Variables

```python
le = LabelEncoder()

df["employment_type"] = le.fit_transform(df["employment_type"])
df["loan_type"] = le.fit_transform(df["loan_type"])
```

## Feature Selection

```python
df = df.drop(columns=["user_id"])
```

## Split Features & Target

```python
X = df.drop("loan_status", axis=1)
y = df["loan_status"]
```

## Scaling

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_scaled = pd.DataFrame(X_scaled, columns=X.columns)
```

## Final Dataset

```python
final_df = X_scaled.copy()
final_df["loan_status"] = y.values

final_df.head()
```

## Save Processed Data

```python
final_df.to_csv("../data/processed/final_dataset.csv", index=False)
```
